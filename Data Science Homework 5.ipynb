```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#1
#I looked through all three of the sources and all of them used data from The Covid Tracking Project.
#At least this is where they cited that most of their data came from so I decided to use that
```

```python
World = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
WorldTotals = World.groupby('date').sum()
WorldTotals['Total Cases'] = WorldTotals['total_cases']
```

```python
States = pd.read_csv('https://covidtracking.com/api/v1/states/daily.csv')
States.date = pd.to_datetime(States.date,format="%Y%m%d")
#States.date=States.date.astype(str)
```

```python
#Q2
WorldTotals.plot(figsize=(30, 10),
    y ='Total Cases',
    title = 'Worldwide Covid Cases by Date'
    )
plt.xlabel("Date")
plt.ylabel("Total Cases (in Tens of Millions)")
```

```python
Countries = World[['location','date','total_cases','new_cases']]
def Country(name):
    Country = Countries[Countries['location'] == name]
    Country['Total Cases'] = Country['total_cases']
    Country['New Cases'] = Country['new_cases']
    Total_Graph = Country.plot(figsize = (20,5),
            title = 'Total Number of Cases in '+name+' by Date',
            y = 'Total Cases',
            x = 'date')
    Total_Graph.set_xlabel("Date")
    Total_Graph.set_ylabel("Total Number of Cases")

    New_Graph = Country.plot(figsize = (20,5),
            title = 'New Cases per Day in '+name,
            y = 'New Cases',
            x = 'date')
    New_Graph.set_xlabel("Date")
    New_Graph.set_ylabel("New Cases")
```

```python
Country('Albania')
Country('Canada')
Country('United Kingdom')
Country('Ireland')
Country('Italy')
Country('South Korea')
Country('Japan')
Country('Montenegro')
```

```python
#3)
#Washinton, Vermont, and New Jersey seem to have flattened the curve

#Grabs all states from data
#bump = States['state']
#bump=bump.unique()
bump = ['WA','VT','NJ']
plt.figure(figsize=(20,5))
for i in range(0,len(bump)):
    state = States[States['state'] == bump[i]]
    plt.plot(state['date'],state['positive'],label=bump[i])
plt.title("Covid-19 Cases by State") 
plt.xlabel("Date")
plt.ylabel("Positive Tests")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=10)
```

```python
#takes variable arguements to compare states from data
#data updates one per week
def StateCompare(*bump):
    plt.figure(figsize=(20,5))
    for i in range(0,len(bump)):
        state = States[States['state'] == bump[i]]
        plt.plot(state['date'],state['positive'],label=bump[i])
    plt.title("Covid-19 Cases by State") 
    plt.xlabel("Date")
    plt.ylabel("Positive Tests")
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=10)
```

```python
StateCompare('WA','FL','RI','CO')
```

```python
Counties = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')
```

```python
#Q4
#Does just Washington Counties
def countyWashington():
    Counties2 = Counties[Counties['state']== 'Washington']
    Counties2 = Counties2.groupby('county').sum()
    Counties2 = Counties2.sort_values('cases')
    Highest = Counties2.iloc[-5:]
    toGraph=[Counties2.loc['Thurston']]
    for x in range(0,len(Highest)):
        toGraph.append(Highest.iloc[x])
    Graphing = pd.DataFrame(toGraph)
    Graphing = Graphing.drop(['fips'], axis=1)
    Graphing.plot.barh()
```

```python
def county(name):
    Counties2 = Counties[Counties['state']== name]
    Counties2 = Counties2.groupby('county').sum()
    Counties2 = Counties2.sort_values('cases')
    Highest = Counties2.iloc[-5:]
    Graphing = Highest.drop(['fips'], axis=1)
    Graphing.plot.barh(figsize = (30,10))
```

```python
#Q4 Answer
countyWashington()
county('South Dakota')
```
